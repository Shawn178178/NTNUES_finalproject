```python
#Midterm1 #Shawn-40844032s #Kaien-40844010s
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from obspy import UTCDateTime

data=pd.read_csv('find.csv',encoding= 'unicode_escape')
d=data["Data"].values
fb=data["fb"].values
FB=data["FB"].values
ig=data["ins"].values
D=np.array(list(map(UTCDateTime,d)))
x=np.linspace(D[0].timestamp,D[-1].timestamp,12)
values=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

#fig
fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(10,6),dpi=80)


ax.plot(D,FB,label='2020Facebook',color='lightskyblue')
ax.plot(D,fb,label='2021facebook',color='navy')
ax.plot(D,ig,color='deeppink',label="instagram")
ax.set_xlabel('Month',fontsize=18)
ax.set_ylabel('Reach',fontsize=18)
ax.set_title('NTNUES Expo Hits',fontsize=25)

#plt.axhline(y=755.5,linestyle='--',color='navy')
#plt.axhline(y=2417.316667,linestyle='--',color='lightseagreen')

plt.grid()
plt.legend(fontsize=16,shadow=True)
plt.xticks(x,values)
fig.tight_layout()
plt.savefig("Hits")
```

```python
#Midterm3 2020 #Shawn-40844032s #Kaien-40844010s
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv('Share.csv',encoding= 'unicode_escape')
shr20=data['2020share'].values
hits20=data['ï»¿2020hits'].values

A=np.array([np.ones(len(shr20)),shr20]).T
m=np.linalg.inv(A.T.dot(A)).dot(A.T).dot(hits20)#斜率矩陣
S=np.linspace(shr20.min(),shr20.max(),81)#回歸線x陣列
H=m[1]*S+m[0]#回歸線y陣列

e=hits20-A.dot(m)
E=(e.T).dot(e)
RR=1-E/np.sum((H-np.mean(H))**2)#最小平方法 

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,6),dpi=80)
ax.plot(shr20,hits20,'o',color='purple',label=r"$Reach={0:8.2f}+{1:8.2f}Share$".format(float(m[0]),float(m[1]))) 
ax.legend()
ax.plot(S,H,'--')
plt.title('2020 Yee',fontsize=18)
plt.xlabel('Share',fontsize=12)
plt.ylabel('Reach',fontsize=12)
ax.text(100,10000,"$R^2$={0:5.2f}".format(float(RR)))

plt.savefig('2020Share.pdf')
```
